```python
# Plotando vetores
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D

def show_image(name, ext = "png"):
    image_path = "/mnt/d/notebook-etc/" + name + "." + ext
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis('off')  # Hide the axes
    plt.show()


u = np.array([2, 5])
v = np.array([5, -2])
w = u + v

# Plot the vectors
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='g', label='u + v')  # Resultant vector


# Set axes limits and labels
plt.xlim(-6, 10)
plt.ylim(-6, 10)
plt.grid()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Vectors u and v')
plt.legend()
plt.show()
```

```python
# Calculando angulo entre dois vetores
cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
```

```python
u = np.array([-7, -2])
v = np.array([8, 3])
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)
# Plot the vectors
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')
# Add text annotations to show the norms
plt.text(u[0] + 0.5, u[1] + 2, f'|u| = {norm_u:.2f}', color='r', fontsize=12)
plt.text(v[0] - 2, v[1] + 0.5, f'|v| = {norm_v:.2f}', color='b', fontsize=12)

# Draw x and y axes (at y=0 and x=0)
plt.axhline(0, color='black',linewidth=1)  # x-axis
plt.axvline(0, color='black',linewidth=1)  # y-axis

# Set axes limits and labels
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Vectors u and v')
plt.legend()
plt.show()
```

```python

# Define the pyramid vertices
s1 = np.array([1, 2, 3])  # Base vertex 1
s2 = np.array([7, 4, 3])  # Base vertex 2
s3 = np.array([2, 1, 9])  # Base vertex 3
apex = np.array([3, 2, 7])  # Apex of the pyramid
d_s1_s2 = np.linalg.norm(s1-s2)
d_s1_s3 = np.linalg.norm(s1-s3)
d_s2_s3 = np.linalg.norm(s2-s3)

# Create a 3D plot
fig = plt.figure(figsize=(12, 8))  # Width 12 inches, Height 8 inches
ax = fig.add_subplot(111, projection='3d')

# Plot the base of the pyramid (connecting the base vertices)
ax.plot([s1[0], s2[0]], [s1[1], s2[1]], [s1[2], s2[2]], color='r')  # Edge between s1 and s2
ax.plot([s2[0], s3[0]], [s2[1], s3[1]], [s2[2], s3[2]], color='r')  # Edge between s2 and s3
ax.plot([s3[0], s1[0]], [s3[1], s1[1]], [s3[2], s1[2]], color='r')  # Edge between s3 and s1


# Plot the edges connecting the apex to each base vertex
ax.plot([apex[0], s1[0]], [apex[1], s1[1]], [apex[2], s1[2]], color='g')  # Edge between apex and s1
ax.plot([apex[0], s2[0]], [apex[1], s2[1]], [apex[2], s2[2]], color='g')  # Edge between apex and s2
ax.plot([apex[0], s3[0]], [apex[1], s3[1]], [apex[2], s3[2]], color='g')  # Edge between apex and s3

# Add labels to each vertex
ax.text(s1[0], s1[1], s1[2], 's1', color='r', fontsize=12, ha='left')
ax.text(s2[0], s2[1], s2[2], 's2', color='r', fontsize=12, ha='left')
ax.text(s3[0], s3[1], s3[2], 's3', color='r', fontsize=12, ha='left')
ax.text(apex[0], apex[1], apex[2], 'top ref', color='g', fontsize=12, ha='left')

#Norms
ax.text(s2[0] - 5, s2[1], s2[2], f'|ds1_s2| = {d_s1_s2:.2f}', color='b', fontsize=12)
ax.text(s3[0], s3[1], s3[0] + s3[2]/2, f'|ds1_s3| = {d_s1_s3:.2f}', color='b', fontsize=12)
ax.text(s2[0]/2, s2[1], s3[2]/1.5, f'|ds2_s2| = {d_s2_s3:.2f}', color='b', fontsize=12)

# Set labels for the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set plot title
ax.set_title('Distances between the satellites')

# Adjust the view to make s3 appear on top
ax.view_init(elev=5, azim=295)

# Show the plot
plt.show()

print(d_s1_s2)
```
